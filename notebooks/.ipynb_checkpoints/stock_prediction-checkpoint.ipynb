```python
import statsmodels.api as sm

# Ensure 'Close' column is a 1-dimensional series for decomposition
data_close = data['Close']['AAPL']  # Adjust to your MultiIndex format

# Decompose the time series
decomposition = sm.tsa.seasonal_decompose(data_close, model='multiplicative', period=252)  # 252 trading days in a year

# Plot the decomposition
plt.rc("figure", figsize=(12, 8))
decomposition.plot()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

# Plot autocorrelation of residuals
plot_acf(decomposition.resid.dropna(), lags=30)  # dropna() to handle any missing values in residuals
plt.title("Autocorrelation of Residuals")
plt.show()
```

```python
# Display the number of missing values per column
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Reset the column index to remove 'Ticker' level
data.columns = data.columns.droplevel(1)  # Removes the 'AAPL' level


Q1 = data['Close'].quantile(0.25)
Q3 = data['Close'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data['Close'] < lower_bound) | (data['Close'] > upper_bound)]
print("Outliers detected:\n", outliers)
```

```python
# Resample to weekly frequency (using the last day of the week)
weekly_data = data['Close'].resample('W').last()
```

```python
# 20-day moving average
data['MA_20'] = data['Close'].rolling(window=20).mean()

# 50-day moving average
data['MA_50'] = data['Close'].rolling(window=50).mean()

# 20-day volatility (standard deviation)
data['Volatility'] = data['Close'].rolling(window=20).std()
```

```python
from scipy import stats

# Calculate Z-scores
data['Z_score'] = stats.zscore(data['Close'])
outliers = data[(data['Z_score'] > 3) | (data['Z_score'] < -3)]
print("Outliers based on Z-score:\n", outliers)
```

```python
# Chosing model and fitting it...
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
plot_acf(data['Close'])
```

```python
f = plt.figure()
ax1= f.add_subplot(121)
ax1.set_title('1st Order Differencing')
ax1.plot(data['Close'].diff())

ax2 = f.add_subplot(122)
plot_acf(data['Close'].diff().dropna(), ax = ax2)
plt.show()
```

```python
f = plt.figure()
ax1= f.add_subplot(121)
ax1.set_title('2nd Order Differencing')
ax1.plot(data['Close'].diff().diff())

ax2 = f.add_subplot(122)
plot_acf(data['Close'].diff().diff().dropna(), ax = ax2)
plt.show()
```

```python
#Choosing order of differentiation:

from statsmodels.tsa.stattools import adfuller
result = adfuller(data['Close'].dropna())
print('p-value:', result[1])

result = adfuller(data['Close'].diff().dropna())
print('p-value:', result[1])

result = adfuller(data['Close'].diff().diff().dropna())
print('p-value:', result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

f = plt.figure()
ax1= f.add_subplot(121)
ax1.set_title('1st Order Differencing')
ax1.plot(data['Close'].diff())

ax2 = f.add_subplot(122)
plot_pacf(data['Close'].diff().dropna(), ax = ax2)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

f = plt.figure()
ax1= f.add_subplot(121)
ax1.set_title('2nd Order Differencing')
ax1.plot(data['Close'].diff().diff())

ax2 = f.add_subplot(122)
plot_pacf(data['Close'].diff().diff().dropna(), ax = ax2)
plt.show()
```

```python
# Set ARIMA model parameters
p, d, q = 1, 1, 1

# Fit the ARIMA model on the differenced data
model = ARIMA(data['Close'], order=(p, d, q))
arima_result = model.fit()

# Print model summary
print(arima_result.summary())
```

```python
# Define number of steps to forecast
forecast_steps = 10

# Forecast the next points
forecast = arima_result.forecast(steps=forecast_steps)
print("Forecasted values:\n", forecast)

# Plot the actual vs. forecasted values
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Actual Close Price')
plt.plot(forecast.index, forecast, color='red', label='Forecasted Close Price')
plt.title('ARIMA Model Forecast vs. Actual')
plt.legend()
plt.show()
```

```python
# Get the last few actual values for comparison
actual_values = data['Close'].iloc[-forecast_steps:]

# Calculate error metrics
mae = mean_absolute_error(actual_values, forecast)
rmse = np.sqrt(mean_squared_error(actual_values, forecast))

print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
P, D, Q, S = 1, 1, 1, 20  # Seasonal ARIMA parameters, with S=5 for weekly seasonality

# Fit the SARIMA model
sarima_model = SARIMAX(data['Close'], order=(p, d, q), seasonal_order=(P, D, Q, S))
sarima_result = sarima_model.fit()

# Print the summary of the SARIMA model
print(sarima_result.summary())
```
